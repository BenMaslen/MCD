```python
import os
import pandas as pd
import glob
```

```python
pest_mon_ims_all = os.listdir("../data/Wildcount/surrounding_images/pest_monitoring/")
```

```python
pest_monitoring_dat_frame = pd.DataFrame({'image_name':list(pest_mon_ims_all)})
```

```python
pest_monitoring_dat_frame.to_csv("../data/Wildcount/location_data/pest_monitoring_images.csv",Index=False)
```

```python
pest_monitoring_dat_frame = pd.read_csv("../data/Wildcount/location_data/pest_monitoring_images.csv")
```

```python
labelled_pm_images = pd.read_csv("../data/Wildcount/location_data/non_wildcount_images.csv")
```

```python
original_image_name = list(
    map(lambda item: "_".join(item.split("_")[:-1])+"."+ item.split('.')[-1], labelled_pm_images['image_names'])
)
```

```python
labelled_pm_images['original_filename'] = original_image_name
```

```python
original_image_name
```

```python
surrounding_images = []

for image in original_image_name:
    if image in list(pest_monitoring_dat_frame['image_name']):
        surrounding_images.append(True)
    else:
        surrounding_images.append(False)
```

```python
labelled_pm_images['labelled'] = surrounding_images
```

```python
Species = []

for image in labelled_pm_images["image_names"]:
    Species.append(glob.glob("../data/Wildcount/training_data/images/*/" + image)[0].split("\\")[1])
```

```python
Species
```

```python
labelled_pm_images['species'] = Species
```

```python
labelled_pm_images.to_csv("../data/Wildcount/location_data/non_wildcount_images.csv")
```
