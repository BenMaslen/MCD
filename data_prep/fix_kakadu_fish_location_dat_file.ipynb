```python
import os
import shutil
```

```python
all_text_files = os.listdir("../data/Kakadu_fish/training_data/labels_all")
```

```python
images_with_labels = list(
    map(lambda item: item.split('.')[0], all_text_files)
)
```

```python
all_image_files = os.listdir("../data/Kakadu_fish/training_data/images_all")
```

```python
all_images = list(
    map(lambda item: item.split('.')[0], all_image_files)
)
```

```python
images_no_labels = []

for im in all_images:
    if im not in images_with_labels:
        images_no_labels.append(im)
```

```python
for image in images_no_labels:
    shutil.copy("../data/Kakadu_fish/training_data/images_all/"+image+".jpg","../data/Kakadu_fish/training_data/images_no_label/"+image+".jpg")
```
