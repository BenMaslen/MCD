```python
import os
import pandas as pd

#vid_locations path
vid_locations_path = "../../data/Deepfish/surrounding_videos/"

#location_data folder path
location_data = "../../data/Deepfish/location_data/"





#get video names
video_name = os.listdir(vid_locations_path)

#get video locations and folder
vid_location = []
vid_folder = []

for video_nam in video_name:
    vid_location.append(vid_locations_path + video_nam)
    vid_folder.append(vid_locations_path)

#get video type    
vid_type = list(
    map(lambda item: item.split('.')[-1], video_name)
)

#get video location
video_location_dat = pd.DataFrame({'video_location':vid_location,'video_name':video_name,'file_type':vid_type,'video_folder':vid_folder})

video_location_dat.to_csv(location_data+"vid_locations.csv",index=False)
```
