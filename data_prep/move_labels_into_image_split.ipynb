```python
import os
import pandas as pd
import shutil
```

```python
test_names_old = os.listdir("../data/Tassy_BRUV/training_data/labels/test/") 
train_names_old = os.listdir("../data/Tassy_BRUV/training_data/labels/train/") 
valid_names_old = os.listdir("../data/Tassy_BRUV/training_data/labels/valid/") 
```

```python
test_names_new = list(
    map(lambda item: item.split('.')[0][:-4]+'.txt', test_names_old)
)
train_names_new = list(
    map(lambda item: item.split('.')[0][:-4]+'.txt', train_names_old)
)
valid_names_new = list(
    map(lambda item: item.split('.')[0][:-4]+'.txt', valid_names_old)
)
```

```python
#for i in range(len(test_names_old)):
#    shutil.move("../data/Tassy_BRUV/training_data/labels/test/"+test_names_old[i],"../data/Tassy_BRUV/training_data/labels/test/"+test_names_new[i])
for i in range(len(train_names_new)):
    shutil.move("../data/Tassy_BRUV/training_data/labels/train/"+train_names_old[i],"../data/Tassy_BRUV/training_data/labels/train/"+train_names_new[i])
for i in range(len(valid_names_new)):
    shutil.move("../data/Tassy_BRUV/training_data/labels/valid/"+valid_names_old[i],"../data/Tassy_BRUV/training_data/labels/valid/"+valid_names_new[i])
```
