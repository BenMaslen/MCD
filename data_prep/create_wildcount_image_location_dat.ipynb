```python
import pandas as pd
import numpy as np
import os
```

```python
wildcount_location_dat = pd.read_csv("../data/Wildcount/location_data/image_location_dat.csv")
```

```python
wildcount_location_dat['filename']
```

```python
original_image_name = list(
    map(lambda item: "_".join(item.split("_")[:-1])+"."+ item.split('.')[-1], wildcount_location_dat['filename'])
)
```

```python
wildcount_location_dat['original_filename'] = original_image_name
```

```python
surrounding_image_location = list(
    map(lambda item: "../../" + "/".join(item.split("/")[3:]), wildcount_location_dat['orgSourceFile'])
)
```

```python
surrounding_image_folder = list(
    map(lambda item: "../../" + "/".join(item.split("/")[3:-1]) + "/", wildcount_location_dat['orgSourceFile'])
)
```

```python
wildcount_location_dat['image_location'] = surrounding_image_location
wildcount_location_dat['image_folder'] = surrounding_image_folder
```

```python
wildcount_location_dat.to_csv("../data/Wildcount/location_data/image_location_dat.csv",index=False)
```
