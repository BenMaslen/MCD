```python
#import libraries
import os
import pandas as pd
import numpy as np

#convert a file name to the appropriate naming convention
dir = "../data/Tassy_BRUV/training_data/images/"
sav_dir = "../data/Tassy_BRUV/image_location/"

test = os.listdir(dir + "test/")
valid = os.listdir(dir + "valid/")
train = os.listdir(dir + "train/")

image_names = [item for sublist in [test,valid,train] for item in sublist]
splits = np.repeat(np.array(["test","valid","train"]), [len(test), len(valid),len(train)], axis=0)

data_split = pd.DataFrame({'image_name':image_names,"split":splits})

data_split.to_csv(sav_dir+"data_split.csv",index=False)
```
