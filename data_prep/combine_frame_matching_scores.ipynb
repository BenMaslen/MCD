```python
#importlibraries

#libraries to get files and move directories
import glob
import shutil
import os


#libraries to extract frame
import pandas as pd
import numpy as np
```

```python
#score directory
score_dir = "../../data/Kakadu_fish/other/matching_images/scores/stage_2_scores/"

#save directory & name
save_dir = "../../data/Kakadu_fish/other/matching_images/scores/stage_2_scores.csv"

#list of scores
scores = os.listdir(score_dir)
```

```python
#load first score
score = pd.read_csv(score_dir + scores[0])

#append the first score with the rest
for i in range(len(scores)):
    if i>0:
        score = score.append(pd.read_csv(score_dir + scores[i])) 
```

```python
score.to_csv(save_dir,index=False)
```
