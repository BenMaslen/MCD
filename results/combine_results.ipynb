```python
#read in libraries
import pandas as pd
import numpy as np
import os

#data set name
data_set = "Tassy_BRUV"
#data_set = "Deepfish"
#data_set = "Kakadu_fish"
#data_set = "Wildcount"

#record directories to save within
results_folder = "../../results/"
data_folder = results_folder + data_set
model_names = os.listdir(data_folder)
```

```python
#go through each folder, look for the summary data set and join them together
mean_results = pd.read_csv(data_folder + '/' + model_names[0] + '/metrics_mean.csv')

mean_results['model_name'] = model_names[0]

for i in range(1,len(model_names)):
    try:
        mean_result_temp = pd.read_csv(data_folder + '/' + model_names[i] + '/metrics_mean.csv')
        mean_result_temp['model_name'] = model_names[i]
        mean_results = mean_results.append(mean_result_temp)
    except:
        pass
```

```python
#save the results
mean_results.to_csv(data_folder + '/all_avg_results.csv',index=False)
```

```python
#go through each folder, look for the summary data set and join them together
mean_results_val = pd.read_csv(data_folder + '/' + model_names[0] + '/metrics_mean_valid.csv')

mean_results_val['model_name'] = model_names[0]

for i in range(1,len(model_names)):
    try:
        mean_result_temp = pd.read_csv(data_folder + '/' + model_names[i] + '/metrics_mean_valid.csv')
        mean_result_temp['model_name'] = model_names[i]
        mean_results_val = mean_results_val.append(mean_result_temp)
    except:
        pass
```

```python
mean_results_val.to_csv(data_folder + '/all_avg_results_val.csv',index=False)
```
