```python
#importlibraries

#libraries to extract frame
import os
import pandas as pd
import cv2
import numpy as np
import shutil
import glob
```

```python
#location data directory
image_to_vid_file = '../../data/Wildcount/location_data/location_dat.csv'

#new location dat
new_image_to_vid_file = '../../data/Wildcount/location_data/location_dat_filt.csv'

#training data directory
training_dat_loc = '../../data/Wildcount/training_data/'

#labels
labels_loc = training_dat_loc + 'labels/'
#images
images_loc = training_dat_loc + 'images/'



image_to_vid_dat = pd.read_csv(image_to_vid_file)
```

```python
#list of all images:
images = [os.listdir(images_loc+'train'),os.listdir(images_loc+'test'),os.listdir(images_loc+'valid')]
images = [item for sublist in images for item in sublist]
```

```python
#list of all labels:
images = [os.listdir(images_loc+'train'),os.listdir(images_loc+'test'),os.listdir(images_loc+'valid')]
images = [item for sublist in images for item in sublist]
```

```python
images_alt = glob.glob(images_loc + '*/*.jpg')
labels_alt = glob.glob(labels_loc + '*/*.txt')
```

```python
images_as_labels = list(
    map(lambda item: item.replace('images','labels').replace(".jpg",".txt"), images_alt)
)
```

```python
non_existant_labels = []

for im_lb in images_as_labels:
    if im_lb not in labels_alt:
        non_existant_labels.append(im_lb)
```

```python
images_without_labels = list(
    map(lambda item: item.replace('labels','images').replace(".txt",".jpg"), non_existant_labels)
)
```

```python
new_image_location = list(
    map(lambda item: item.replace('images','images_no_labels'), images_without_labels)
)
```

```python
image_names_without_labels = list(
    map(lambda item: item.split("\\")[-1], images_without_labels)
)
```

```python
row_has_lbl = []

for im in image_to_vid_dat['image_name']:
    if im in image_names_without_labels:
        row_has_lbl.append("yes")
    else:
        row_has_lbl.append("no")
```

```python
image_to_vid_dat['has_label'] = row_has_lbl
```

```python
image_to_vid_dat.to_csv(new_image_to_vid_file,index=False)
```

```python
#move images without labels
im_nums = range(len(images_without_labels))

for im_num in im_nums: 
    shutil.move(images_without_labels[im_num],new_image_location[im_num])
```
